```python
import pandas as pd
import plotly.express as px
```

```python
data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salário': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}

df = pd.DataFrame(data)
df.head(5)
```

```python
# Mapa de arvore de estado civil salario e limite de credito

# Agrupando os dados pela média do limite para não distorcer o tamanho das caixas
df_agrupado = df.groupby(['Estado_Civil', 'Profissão'], as_index=False)['Limite_Credito'].mean()

fig = px.treemap(
    df_agrupado,
    path=[px.Constant("Todos"), 'Estado_Civil', 'Profissão'],
    values='Limite_Credito',
    color='Limite_Credito',
    title='Média de Limite de Crédito por Estado Civil e Profissão'
)

fig.show()
```

Quis gerar um mapa de arvore por 2 motivos, 1 era um dos mais frescos na minha cabeça, e 2 poderia identificar diversos dados de uma so vez.
Foi possivel perceber que ter uma profissão era importante, ja que o estudante é o com o menor limite de credito entre todos, e que ser casado é tão importante quanto, ja que a media de limite de credito dos casados é o dobro do grupo de solteiros  

```python
# Grafico de barras para verificar importancia do historico de inadimplencia e imovel proprio no limite de credito
df_medias = df.groupby(['Historico_Inadimplencia', 'Imovel_Proprio'])['Limite_Credito'].mean().reset_index()

fig_bar = px.bar(df_medias, 
                 x="Historico_Inadimplencia", 
                 y="Limite_Credito", 
                 color="Imovel_Proprio", 
                 barmode='group',
                 text_auto='.2s',
                 title="Média do Limite de Crédito por Inadimplência e Imóvel Próprio",
                 labels={"Historico_Inadimplencia": "Histórico de Inadimplência", 
                         "Limite_Credito": "Média do Limite de Crédito (R$)",
                         "Imovel_Proprio": "Possui Imóvel?"})
fig_bar.show()
```

Depois quis verificar a importancia do historico de inadimplência e do imovel proprio para o limite, como ja era esperado ter um imóvel proprio afeta bastante no limite de credito, além que os que estão em inadimplência estão com o limite muito abaixo.

```python
# Grafico de dispersão para entender idade e limite de credito
fig = px.scatter(df, x='Idade', y='Limite_Credito', color='Profissão', title='Limite de Crédito por Idade')
fig.show()
```

Para finalizar gostaria de entender se a idade seria um fator determinante para o limite de credito, aparentemente abaixo dos 30 anos realmente a idade faz diferença ja que o candidato com 25 anos mesmo tendo uma profissão ainda tem o limite bem abaixo.
